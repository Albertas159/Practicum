## Analysis and cropping

```python
df["book_wordcount"]= df["full_text"].map(lambda x: len(x.split(" ")))
df["sum_wordcount"]= df["summary"].map(lambda x: len(x.split(" ")))
df["ratio"]= df["sum_wordcount"]*100/df["book_wordcount"]
```

```python
df.describe()
```

```python
# Ratio= 100 * book word count / summary word count
# Summs more than 0.1% length of books
df_trunc = df.loc[df["ratio"]>0.1]
```

```python
# Summs less than 5% length of books
df_trunc = df_trunc.loc[df["ratio"]<10]
```

```python
# Book length less than 200k words
df_trunc = df_trunc.loc[df["book_wordcount"]<200000]
```

```python
# Book summary more than 100 words
df_trunc = df_trunc.loc[df["sum_wordcount"]>100]
```

```python
df = df_trunc[["title","full_text","summary","t5_sum"]]
```

```python
df = df.reset_index(drop=True)
```
